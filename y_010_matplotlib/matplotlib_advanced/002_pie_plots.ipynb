## importing data and cleaning it

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast

#loading data set
dataset1 = load_dataset('lukebarousse/data_jobs')
dataf = dataset1['train'].to_pandas()
df = dataf.copy()

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

### First

```python
df['job_work_from_home'].value_counts()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',colormap='summer')
plt.title('Job_WOrk_fromhome')
plt.ylabel(' ')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='pie', colormap='pink')
plt.title('Job_Titles')
plt.ylabel(' ')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='pie', colormap='summer',autopct='%1.1f%%')
plt.title('Job_Titles')
plt.ylabel(' ')
plt.show()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', colormap='summer',startangle=180,autopct='%1.1f')
plt.title('Job_WOrk_fromhome')
plt.ylabel(' ')
plt.show()
```

```python
df['job_work_from_home'].value_counts().plot(
    kind='pie', colormap='summer', startangle=180, autopct='%1.1f%%')
plt.title('Job_WOrk_fromhome')
plt.ylabel(' ')
plt.show()
```

### Second

```python
df[['job_work_from_home','job_no_degree_mention','job_health_insurance']]
```

```python
fig,ax=plt.subplots(1,3)
dict_colu = {
    'job_work_from_home':'Work from home',
    'job_no_degree_mention':'job_degree',
    'job_health_insurance':'job_health'
}

for i,(columns,titles) in enumerate(dict_colu.items()):
    ax[i].pie(df[columns].value_counts(),autopct='%1.1f%%',startangle=180,labels=['False','True'])
    ax[i].set_title(titles)

plt.show()
```
