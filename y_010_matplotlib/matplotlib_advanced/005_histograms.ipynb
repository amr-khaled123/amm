## Imorting and Cleaning data

```python
import matplotlib.pyplot as plt
import pandas as pd
import ast
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
data = dataset['train'].to_pandas()
df = data.copy()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

### Data Frame of Data Anlyst of USA

```python
df_DA_US = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')].copy()
```

```python
df_DA_US['salary_year_avg'].plot(kind='hist',bins=40,edgecolor='yellow',color='red')
plt.xlim(0,250000)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x,pos:f'${int(x/1000)}K'))

plt.title('United States Yearly Salary of Data Analyst')
plt.ylabel('number of jobs')
plt.xlabel('Yearly Salary')
plt.show()
```

### Data Frame of Data Scientist of USA

```python
df_DS_US = df[(df['job_title_short']=='Data Scientist')&(df['job_country']=='United States')].copy()
```

```python
df_DS_US['salary_year_avg'].plot(kind='hist',bins=40,edgecolor='black')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x,_:f"${int(x/1000)}K"))
plt.xlim(0,250000)

plt.title('United States Yearly Salary of Data Scientist')
plt.ylabel('number of jobs')
plt.xlabel('Yearly Salary')
plt.show()
```
