```python
job_counts = df['job_title_short'].value_counts()
```

```python
plt.bar(job_counts.index,job_counts)
plt.xlabel('Job_Title')
plt.xticks(rotation = 45,ha='right')
plt.title('jobs visualization')
plt.show()
```

```python
plt.barh(job_counts.index, job_counts)
plt.ylabel('Job_Title')
plt.xticks(rotation=45,ha='right')
plt.title('jobs visualization')
plt.show()
```

```python
job_counts = job_counts.sort_values(ascending=True)
```

```python
plt.barh(job_counts.index, job_counts)
plt.ylabel('Job_Title')
plt.xticks(rotation=90)
plt.title('jobs visualization')
plt.show()
```

```python
plt.pie(job_counts, labels=job_counts.index, autopct='%1.2f%%')
plt.title('My Customized Pie Chart', fontsize=14, color='red')
plt.show()
```
