```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
# loading data set
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.job_posted_date.dt.month_name().value_counts()
```

```python
job_location_count = df['job_country'].value_counts()
job_location_count.head(5)
```

```python
job_location_count.head(5).plot(kind='bar')
plt.title('Job_Location')
plt.xlabel('Countries')
plt.xticks(rotation=45,ha='right')
plt.show()
```
