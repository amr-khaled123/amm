```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
from datasets import load_dataset

dataset1 = load_dataset('lukebarousse/data_jobs')
dataf = dataset1['train'].to_pandas()
df = dataf.copy()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df[['job_title_short','job_skills']]
```

```python
df2 = df.copy()
```

```python
df2['job_title_short'] = df2['job_title_short'].apply(lambda x:x  if 'Senior' in x else None)
```

```python
df2['job_title_short1'] = df2['job_title_short']
```

```python
df2.reset_index(inplace=True)
df2.set_index('job_title_short1',inplace=True)
```

```python
df2 = df2.dropna(subset='job_title_short')
```

```python
df2.reset_index(inplace=True)
```

```python
df2
```

```python
df3 = df.copy()
```

```python
df3
```

```python
df3_explode = df3.explode('job_skills')
```

```python
df3_explode['job_skills'].value_counts().head(10).plot(kind='barh', color=['red','black'])
```

```python
colors = ['red', 'black', 'green','yellow','blue','purple','pink','orange','brown','gray']
```

```python
df3_explode['job_skills'].value_counts().head(10).plot(kind='pie', autopct='%1.1f%%',color=colors,startangle=0)
```
