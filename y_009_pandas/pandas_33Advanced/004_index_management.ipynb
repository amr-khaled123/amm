```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data set
dataset1 = load_dataset('lukebarousse/data_jobs')
df = dataset1['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_1 = df.copy()
df_1
```

```python
a = df_1.loc[0:, 'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
#a['salary_year_avg'].sort_values()
a.dropna(subset='salary_year_avg')
#a.dropna(subset = 'salary_hour_avg')
a['salary_year_avg'].mean()
```

```python
df.index.name = 'job_index'
df
```

```python
df_ind = df[df['job_country']=='India']
df_ind=df_ind.loc[0:].dropna(subset='salary_rate')
df_ind=df_ind.loc[0:].dropna(subset='salary_year_avg')
```

```python
df_ind.reset_index(inplace=True)
df_ind
```

```python
df_ind.set_index('job_index',inplace=True)
```

```python
df_ind
```

```python
median_pivot = df_ind.pivot_table(values='salary_year_avg',index='job_title_short',aggfunc=['max','median','mean','min'])
median_pivot
```

```python
median_pivot.sort_index()
```

```python
median_pivot.plot(kind='bar')
plt.title('software india jobs')
plt.xlabel('job_title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
