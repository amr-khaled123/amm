```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
dataset1 = load_dataset('lukebarousse/data_jobs')
df = dataset1['train'].to_pandas()
df1 = df.copy()
df1['job_posted_date'] = pd.to_datetime(df1['job_posted_date'])
```

```python
df1.groupby('job_title_short').describe()
```

```python
df1.groupby('job_title_short').agg({'job_posted_date':['count'],'salary_year_avg':['min','mean','median']}).plot(kind='bar')
plt.title('Job_title_short describe')
plt.ylim(0,200000)
plt.xlabel('jobs')
plt.xticks(rotation=45,ha='right')
```
